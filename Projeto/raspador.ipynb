```python
import requests
from bs4 import BeautifulSoup

# Lista de URLs das páginas que você deseja raspar
urls = [
    "https://diariooficialdasprefeituras.org/piaui/buscas?page=1",
    "https://diariooficialdasprefeituras.org/piaui/buscas?page=2",
    "https://diariooficialdasprefeituras.org/piaui/buscas?page=3",
    # Adicione mais URLs, se necessário
]

# Inicialize o nome base do arquivo
base_file_name = "licitaçao"

# Estrutura HTML base
base_html = """<!DOCTYPE html>
<html lang="pt-br">

<head>
    <meta charset="utf-8" />
    <meta content="width=device-width, initial-scale=1, shrink-to-fit=no" name="viewport" />
    <link href="licitaçao.css" rel="stylesheet" /> <!-- Importa um arquivo de estilos personalizados -->
    <link crossorigin="anonymous" href="https://stackpath.bootstrapcdn.com/bootstrap/4.1.3/css/bootstrap.min.css"
        integrity="sha384-MCw98/SFnGE8fJT3GXwEOngsV7Zt27NXFoaoApmYm81iuXoPkFOJwJ8ERdknLPMO" rel="stylesheet" />
    <!-- Importa a biblioteca de estilos do Bootstrap -->
    <title>Diario Oficial - Licitações</title> <!-- Define o título da página -->
</head>

<body>
    <nav class="navbar navbar-expand-lg navbar navbar-dark bg-dark">
        <a class="navbar-brand" href="index.html">Acesse Diario</a>
        <button aria-controls="navbarNav" aria-expanded="false" aria-label="Alterna navegação" class="navbar-toggler"
            data-target="#navbarNav" data-toggle="collapse" type="button">
            <span class="navbar-toggler-icon"></span>
        </button>
    </nav>

    <!-- Adicione um formulário de filtro -->
    <div class="container mt-4">
        <form>
            <div class="form-group">
                <label for="palavraChave">Palavra Chave</label>
                <input type="text" class="form-control" id="palavraChave" placeholder="Digite uma palavra-chave">
            </div>
            <div class="form-group">
                <label for="numEdicao">Nº da Edição</label>
                <input type="text" class="form-control" id="numEdicao" placeholder="Digite o número da edição">
            </div>
            <div class="form-group">
                <label for="periodoPublicacao">Período de Publicação</label>
                <input type="text" class="form-control" id="periodoPublicacao" placeholder="Selecione o período de publicação">
            </div>
            <button type="submit" class="btn btn-primary">Filtrar</button>
        </form>
    </div>
    

    <table class="table">
        <thead>
            <!-- Cabeçalho da tabela -->
        </thead>
        <tbody>
            <!-- Linhas da tabela com dados vão aqui -->
        </tbody>
    </table>

    <nav aria-label="Page navigation">
        <ul class="pagination">
            <li class="page-item disabled">
                <a aria-disabled="true" class="page-link" href="#" tabindex="-1">Anterior</a>
            </li>
            <li aria-current="page" class="page-item active">
                <a class="page-link" href="licitaçao1.html">1</a>
            </li>
            <li class="page-item">
                <a class="page-link" href="licitaçao2.html">2</a>
            </li>
            <li class="page-item">
                <a class="page-link" href="licitaçao3.html">3</a>
            </li>
            <li class="page-item">
                <a class="page-link" href="#">Próxima</a>
            </li>
        </ul>
    </nav>

    <!-- JavaScript (Opcional) -->
    <script crossorigin="anonymous" integrity="sha384-q8i/X+965DzO0rT7abK41JStQIAqVgRVzpbzo5smXKp4YfRvH+8abtTE1Pi6jizo"
        src="https://code.jquery.com/jquery-3.3.1.slim.min.js"></script>
    <script crossorigin="anonymous" integrity="sha384-ZMP7rVo3mIykV+2+9J3UJ46jBk0WLaUAdn689aCwoqbBJiSnjAK/l8WvCWPIPm49"
        src="https://cdnjs.cloudflare.com/ajax/libs/popper.js/1.14.3/umd/popper.min.js"></script>
    <script crossorigin="anonymous" integrity="sha384-ChfqqxuZUCnJSK3+MXmPNIyE6ZbWh2IMqE241rYiqJxyMiZ6OW/JmZQ5stwEULTy"
        src="https://stackpath.bootstrapcdn.com/bootstrap/4.1.3/js/bootstrap.min.js"></script>
        <script src="filter.js"></script>

    </body>

</html>


    """

# Loop através das URLs
for index, url in enumerate(urls, start=1):
    # Fazer uma solicitação HTTP para a página
    response = requests.get(url)

    # Verificar se a solicitação foi bem-sucedida
    if response.status_code == 200:
        # Parse o conteúdo HTML da página
        soup = BeautifulSoup(response.text, "html.parser")

        # Localize a tabela que você deseja raspar (você precisará inspecionar o código-fonte HTML para isso)
        table = soup.find("table")

        # Verifique se a tabela foi encontrada
        if table:
            # Crie um nome de arquivo exclusivo para cada página
            file_name = f"{base_file_name}{index}.html"

            # Crie uma cópia da estrutura HTML base
            current_html = base_html

            # Adicione os dados da tabela à estrutura HTML
            current_html += str(table)

            # Salve o arquivo HTML completo
            with open(file_name, "w", encoding="utf-8") as file:
                file.write(current_html)
        else:
            print(f"Tabela não encontrada na página {url}")
    else:
        print(f"Erro ao acessar a página {url}. Código de status: {response.status_code}")

# Feche a estrutura HTML
base_html += """
    </body>
</html>
"""

# Salve a estrutura HTML base em um arquivo
with open("licitaçao.html", "w", encoding="utf-8") as file:
    file.write(base_html)
```
