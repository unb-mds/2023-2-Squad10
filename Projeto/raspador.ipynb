```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# Lista de URLs das páginas que você deseja raspar
urls = [
    "https://diariooficialdasprefeituras.org/piaui/buscas?page=1",
    "https://diariooficialdasprefeituras.org/piaui/buscas?page=2",
    "https://diariooficialdasprefeituras.org/piaui/buscas?page=3",
    # Adicione mais URLs, se necessário
]

# Inicialize o nome base do arquivo
base_file_name = "licitaçao"

# Estrutura HTML base
base_html = """

    
    
        
    
        <!DOCTYPE html>
<html lang="pt-br">

<head>
    <meta charset="utf-8" />
    <meta content="width=device-width, initial-scale=1, shrink-to-fit=no" name="viewport" />
    <link href="licitaçao.css" rel="stylesheet" />
    <link crossorigin="anonymous" href="https://stackpath.bootstrapcdn.com/bootstrap/4.1.3/css/bootstrap.min.css"
        integrity="sha384-MCw98/SFnGE8fJT3GXwEOngsV7Zt27NXFoaoApmYm81iuXoPkFOJwJ8ERdknLPMO" rel="stylesheet" />
    <title>Diario Oficial - Licitações</title>
</head>

<body>
    <nav class="navbar navbar-expand-lg navbar navbar-dark bg-dark">
        <a class="navbar-brand" href="index.html">Acesse Diario</a>
        <button aria-controls="navbarNav" aria-expanded="false" aria-label="Alterna navegação" class="navbar-toggler"
            data-target="#navbarNav" data-toggle="collapse" type="button">
            <span class="navbar-toggler-icon"></span>
        </button>
    </nav>


    <table class="table mt-4">
        <thead>

        </thead>
        <tbody>
            <!-- Linhas da tabela com dados vão aqui -->
        </tbody>
    </table>

    <nav aria-label="Navegação de página exemplo" class="pagination">
        <ul class="pagination justify-content-center">
            <li class="page-item disabled">
                <a class="page-link" href="#" tabindex="-1">Anterior</a>
            </li>
            <li class="page-item"><a class="page-link" href="licitaçao1.html">1</a></li>
            <li class="page-item"><a class="page-link" href="licitaçao2.html">2</a></li>
            <li class="page-item"><a class="page-link" href="licitaçao3.html">3</a></li>
            <li class="page-item">
                <a class="page-link" href="#">Próximo</a>
            </li>
        </ul>
    </nav>

 <script>
    document.addEventListener("DOMContentLoaded", function () {
    document.querySelector(".btn").addEventListener("click", function (event) {
        event.preventDefault(); // Impede o envio do formulário e recarregamento da página

        // Obtenha os valores dos campos de filtro
        var palavraChave = document.getElementById("palavra-chave").value.toLowerCase();
        var identificador = document.getElementById("identificador").value.toLowerCase();
        var numeroEdicao = document.getElementById("numero-edicao").value.toLowerCase();
        var classificacaoAto = document.getElementById("classificacao-ato").value.toLowerCase();
        var entidade = document.getElementById("entidade").value.toLowerCase();
        var dataInicial = document.getElementById("data-inicial").value;
        var dataFinal = document.getElementById("data-final").value;

        // Percorra as linhas da tabela
        var linhasTabela = document.getElementById("dadosTabela").getElementsByTagName("tr");
        for (var i = 0; i < linhasTabela.length; i++) {
            var linha = linhasTabela[i];

            // Se a linha for o cabeçalho, pule para a próxima iteração
            if (linha.cells.length === 0) {
                continue;
            }

            var celulas = linha.getElementsByTagName("td");

            // Obtenha os valores das células relevantes
            var valorPalavraChave = celulas[1].textContent.toLowerCase();
            var valorIdentificador = celulas[2].textContent.toLowerCase();
            var valorNumeroEdicao = celulas[0].textContent.toLowerCase();
            var valorClassificacaoAto = celulas[5].textContent.toLowerCase();
            var valorEntidade = celulas[4].textContent.toLowerCase();
            var valorData = celulas[3].textContent;

            // Aplicar lógica de filtro
            var atendeFiltro = (
                (!palavraChave || valorPalavraChave.includes(palavraChave)) &&
                (!identificador || valorIdentificador.includes(identificador)) &&
                (!numeroEdicao || valorNumeroEdicao.includes(numeroEdicao)) &&
                (!classificacaoAto || valorClassificacaoAto.includes(classificacaoAto)) &&
                (!entidade || valorEntidade.includes(entidade)) &&
                (!dataInicial || (valorData >= dataInicial && valorData <= dataFinal))
            );

            // Ocultar ou mostrar a linha com base no filtro
            linha.style.display = atendeFiltro ? "" : "none";
        }
    });
});



    </script>

    <script crossorigin="anonymous" integrity="sha384-q8i/X+965DzO0rT7abK41JStQIAqVgRVzpbzo5smXKp4YfRvH+8abtTE1Pi6jizo"
        src="https://code.jquery.com/jquery-3.3.1.slim.min.js"></script>
    <script crossorigin="anonymous" integrity="sha384-ZMP7rVo3mIykV+2+9J3UJ46jBk0WLaUAdn689aCwoqbBJiSnjAK/l8WvCWPIPm49"
        src="https://cdnjs.cloudflare.com/ajax/libs/popper.js/1.14.3/umd/popper.min.js"></script>
    <script crossorigin="anonymous" integrity="sha384-ChfqqxuZUCnJSK3+MXmPNIyE6ZbWh2IMqE241rYiqJxyMiZ6OW/JmZQ5stwEULTy"
        src="https://stackpath.bootstrapcdn.com/bootstrap/4.1.3/js/bootstrap.min.js"></script>
    <script src="filter.js"></script>
</body>

</html>

    


    """

# Loop através das URLs
for index, url in enumerate(urls, start=1):
    # Fazer uma solicitação HTTP para a página
    response = requests.get(url)

    # Verificar se a solicitação foi bem-sucedida
    if response.status_code == 200:
        # Parse o conteúdo HTML da página
        soup = BeautifulSoup(response.text, "html.parser")

        # Localize a tabela que você deseja raspar (você precisará inspecionar o código-fonte HTML para isso)
        table = soup.find("table")

        # Verifique se a tabela foi encontrada
        if table:
            # Substitua "/assets/pdf.png" por "/Projeto/assets/pdf.png" nos links de imagem
            for img in table.find_all("img"):
                if "/assets/pdf.png" in img["src"]:
                    img["src"] = img["src"].replace(
                        "/assets/pdf.png", "/Projeto/assets/pdf.png"
                    )

            # Adicione o prefixo "https://diariooficialdasprefeituras.org/" aos links encontrados na classe "lnkDetail"
            for link in table.find_all("a", class_="lnkDetail"):
                if link.has_attr("href"):
                    link["href"] = urljoin(
                        "https://diariooficialdasprefeituras.org/", link["href"]
                    )

            # Adicione o prefixo "https://diariooficialdasprefeituras.org/" aos links encontrados na classe "fileLink"
            for link in table.find_all("a", class_="fileLink"):
                if link.has_attr("data-href"):
                    link["data-href"] = urljoin(
                        "https://diariooficialdasprefeituras.org/", link["data-href"]
                    )

            # Adicione o trecho para substituir a estrutura específica de <a>
            for link in table.find_all(
                "a", class_="fileLink", href="javascript:void(0)"
            ):
                if link.has_attr("data-href"):
                    new_link = soup.new_tag(
                        "a", href=link["data-href"], target="_blank"
                    )
                    new_link.string = "Baixar"
                    link.replace_with(new_link)

            # Crie um nome de arquivo exclusivo para cada página
            file_name = f"{base_file_name}{index}.html"

            # Crie uma cópia da estrutura HTML base
            current_html = base_html

            # Adicione os dados da tabela à estrutura HTML
            current_html += str(table)

            # Salve o arquivo HTML completo
            with open(file_name, "w", encoding="utf-8") as file:
                file.write(current_html)
        else:
            print(f"Tabela não encontrada na página {url}")
    else:
        print(
            f"Erro ao acessar a página {url}. Código de status: {response.status_code}"
        )

# Salve a estrutura HTML base em um arquivo
with open("licitaçao.html", "w", encoding="utf-8") as file:
    file.write(base_html)
```
